# 05.2 K-Nearest Neighbors (KNN) - Optimized

This notebook focuses on **Optimizing** the K-Nearest Neighbors (KNN) classifier using hyperparameter tuning. We aim to find the best value for $K$, the distance metric, and the weighting function to improve the model's accuracy and stability.

```python
from pathlib import Path
import sys
import json
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Add project root to sys.path to import local utilities
PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import utils
```

## 1. Load Processed Data

```python
X_train, X_test, y_train, y_test = utils.load_processed_data()

print("Feature matrix shapes:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

## 2. Train and Evaluate Optimized KNN

We use `GridSearchCV` to tune the following hyperparameters:
- `n_neighbors`: The number of neighbors to use.
- `weights`: Weight function used in prediction ('uniform' or 'distance').
- `metric`: Distance metric to use ('euclidean' or 'manhattan').

```python
knn = KNeighborsClassifier()

param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11, 15],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

metrics, summary_df = utils.train_and_evaluate(
    model=knn, 
    model_name="KNN Optimized", 
    param_grid=param_grid
)

print("Optimal Hyperparameters:")
print(json.dumps(metrics["Best Parameters"], indent=2))

display(summary_df)
```

## 3. Results Visualization

The confusion matrix below shows the final performance of our optimized model.

```python
from IPython.display import Image

print("Classification Report:")
print(pd.DataFrame(metrics["Classification Report"]).transpose().round(4))

print("\nConfusion Matrix:")
display(Image(filename=metrics["Confusion Matrix Figure"]))
```
