# 05.1 K-Nearest Neighbors (KNN) - Baseline

This notebook implements the **Baseline** K-Nearest Neighbors (KNN) algorithm using default parameters ($K=5$ and Euclidean distance). This serves as our starting point for comparing with the optimized version.

```python
from pathlib import Path
import sys
import json
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Add project root to sys.path to import local utilities
PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import utils
```

## 1. Load Processed Data

```python
X_train, X_test, y_train, y_test = utils.load_processed_data()

print("Feature matrix shapes:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

## 2. Train and Evaluate Baseline KNN

We use the default `KNeighborsClassifier` settings, which includes $K=5$ and the **Euclidean** distance metric.

```python
knn_baseline = KNeighborsClassifier()

metrics, summary_df = utils.train_and_evaluate(
    model=knn_baseline, 
    model_name="KNN Baseline"
)

display(summary_df)
```

## 3. Results Visualization

The confusion matrix below shows the initial performance of the baseline model.
