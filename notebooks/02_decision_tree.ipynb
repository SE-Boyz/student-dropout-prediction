# Decision Tree Classifier - Student Dropout Prediction

This notebook implements a **Decision Tree** algorithm to predict student dropout and academic success. We will use `GridSearchCV` to find the optimal hyperparameters for the classifier and evaluate its performance using standardized metrics.

```python
import sys
from pathlib import Path

# Ensure the project root is in the Python path for utils import
PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import utils

# Ensure necessary directories exist
utils.ensure_directories()
```

## 1. Load Data

We load the pre-split and processed datasets from the `data/` directory.

```python
X_train, X_test, y_train, y_test = utils.load_processed_data()

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

## 2. Hyperparameter Tuning

We define a grid of hyperparameters to optimize the model performance using 5-fold cross-validation.

```python
param_grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [5, 10, 15, 20, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "random_state": [42]
}

dt_model = DecisionTreeClassifier(random_state=42)

metrics, summary_df = utils.train_and_evaluate(
    model=dt_model, 
    model_name="Decision Tree", 
    param_grid=param_grid
)
```

## 3. Results Overview

Let's look at the best parameters and the final performance metrics on the test set.

```python
print("Best Parameters Found:")
for param, value in metrics["Best Parameters"].items():
    print(f" - {param}: {value}")

display(summary_df)
```
