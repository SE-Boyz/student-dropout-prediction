# 02 K-Nearest Neighbors (KNN)

This notebook implements the K-Nearest Neighbors algorithm for predicting student dropout and academic success. It uses the preprocessed data generated in the previous step and performs hyperparameter tuning to find the optimal configuration.

```python
from pathlib import Path
import sys
import json
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Add project root to sys.path to import local utilities
PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import utils
```

## Load Processed Data

```python
X_train, X_test, y_train, y_test = utils.load_processed_data()

print("Feature matrix shapes:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

## Train and Evaluate KNN

We will tune the following hyperparameters:
- `n_neighbors`: The number of neighbors to use.
- `weights`: Weight function used in prediction ('uniform' or 'distance').
- `metric`: Distance metric to use for the tree ('euclidean' or 'manhattan').

```python
knn = KNeighborsClassifier()

param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11, 15],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

metrics, summary_df = utils.train_and_evaluate(
    model=knn, 
    model_name="KNN", 
    param_grid=param_grid
)

print("Optimal Hyperparameters:")
print(json.dumps(metrics["Best Parameters"], indent=2))

display(summary_df)
```

## Results Visualization

The confusion matrix provides a detailed view of how the model performs across different classes.

```python
from IPython.display import Image

print("Classification Report:")
print(pd.DataFrame(metrics["Classification Report"]).transpose().round(4))

print("\nConfusion Matrix:")
display(Image(filename=metrics["Confusion Matrix Figure"]))
```
