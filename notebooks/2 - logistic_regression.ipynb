# Logistic Regression Model

This notebook trains and evaluates a Logistic Regression model for the student dropout prediction task. It utilizes the utility functions from `utils.py` for data loading, model evaluation, and artifact tracking.

```python
import sys
import os

# Add parent directory to path to allow importing from utils.py
sys.path.append(os.path.abspath('..'))

from sklearn.linear_model import LogisticRegression
from utils import train_and_evaluate, load_processed_data
import pandas as pd
```

```python
# Load preprocessed data using the utility function
# This will trigger `preprocess_and_save()` if the processed CSVs are not found in the `data/` directory.
X_train, X_test, y_train, y_test = load_processed_data()

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

```python
# Initialize the logistic regression model
# max_iter is increased to ensure convergence for this dataset
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Optional: Define a parameter grid for hyperparameter tuning using GridSearchCV
# param_grid = {
#     'C': [0.1, 1.0, 10.0],
#     'penalty': ['l2']
# }

print("Training the Logistic Regression model...")
metrics, metrics_df = train_and_evaluate(
    model=log_reg, 
    model_name="Logistic Regression"
    # param_grid=param_grid  # Uncomment to enable GridSearch tuning
)

# Display summary performance dataframe
metrics_df
```

```python
from IPython.display import Image, display

# Display standard classification metrics summary
print("\nClassification Report:")
for label, summary in metrics['Classification Report'].items():
    if isinstance(summary, dict):
        print(f"{label}: Precision: {summary['precision']:.4f}, Recall: {summary['recall']:.4f}, F1: {summary['f1-score']:.4f}")
    else:
        print(f"{label}: {summary:.4f}")

# The train_and_evaluate function automatically generates and saves a confusion matrix figure.
print("\nConfusion Matrix (Test Set):")
display(Image(filename=metrics['Confusion Matrix Figure']))
```
