# 01 Preprocessing


## Objective
            This notebook loads the UCI **Predict Students' Dropout and Academic Success** dataset, inspects
            its structure, shows the class distribution, and applies the shared
            preprocessing pipeline used by every model notebook.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

```python
import json
import pandas as pd
from utils import (
    DATASET_DOI,
    DATASET_URL,
    inspect_dataset,
    load_student_dataset,
    preprocess_and_save,
)

df, dataset_path = load_student_dataset()
print(f"Loaded dataset from: {dataset_path}")
print(f"Dataset URL: {DATASET_URL}")
print(f"DOI: {DATASET_DOI}")
print(f"Dataset shape: {df.shape}")

inspection = inspect_dataset(df)
display(df.head())
display(inspection["feature_overview"].head(20))
display(inspection["data_types"].head(20))
display(inspection["missing_values"][inspection["missing_values"]["missing_count"] > 0])
display(inspection["class_distribution"])
display(inspection["numeric_summary"].head(10))

artifacts = preprocess_and_save(test_size=0.2, random_state=42)
metadata_path = PROJECT_ROOT / "results" / "preprocessing_metadata.json"
metadata = json.loads(metadata_path.read_text(encoding="utf-8"))

print("Processed feature matrix shapes:")
print(f"X_train: {artifacts['X_train'].shape}")
print(f"X_test: {artifacts['X_test'].shape}")
print(f"y_train: {artifacts['y_train'].shape}")
print(f"y_test: {artifacts['y_test'].shape}")
print("Saved preprocessing metadata:")
print(json.dumps(metadata, indent=2))

display(artifacts["X_train"].head())
```
