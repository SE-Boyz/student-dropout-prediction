# 02 Random Forest Classification

## Objective
This notebook trains and evaluates a **Random Forest Classifier** for the student dropout prediction task. It reuses the shared preprocessing and evaluation utilities from the project so that results are consistent with the rest of the assignment.

## 1. Import Libraries and Utilities

We import the required libraries and shared project utilities used to train, evaluate, and save the baseline random forest model.

```python
from pathlib import Path
import sys

import pandas as pd
from sklearn.ensemble import RandomForestClassifier

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from utils import METRICS_DIR, MODELS_DIR, FIGURES_DIR, train_and_evaluate
```

## 2. Train Baseline Random Forest

This block defines the baseline random forest configuration and performs a small grid search to find a solid starting set of hyperparameters.

```python
random_forest = RandomForestClassifier(
    random_state=42,
    n_jobs=1,
    class_weight="balanced",
)

param_grid = {
    "n_estimators": [200, 300],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
}

metrics, summary_df = train_and_evaluate(
    model=random_forest,
    model_name="Random Forest",
    param_grid=param_grid,
)

summary_df
```

## 3. Results Overview

We summarize the main evaluation metrics, the selected hyperparameters, and the saved artifact locations for the baseline model.

```python
pd.Series({
    "Accuracy": metrics["Accuracy"],
    "Balanced Accuracy": metrics["Balanced Accuracy"],
    "Weighted Precision": metrics["Precision (weighted)"],
    "Weighted Recall": metrics["Recall (weighted)"],
    "Weighted F1": metrics["F1-score (weighted)"],
    "Macro F1": metrics["F1-score (macro)"],
    "Best Parameters": metrics.get("Best Parameters", {}),
    "Best CV Weighted F1": metrics.get("Best CV Weighted F1"),
    "Metrics JSON": str(METRICS_DIR / "random_forest.json"),
    "Model File": str(MODELS_DIR / "random_forest.joblib"),
    "Confusion Matrix Figure": str(FIGURES_DIR / "random_forest_confusion_matrix.png"),
})
```

## 4. Classification Report

This report shows class-level precision, recall, and F1-score so we can inspect how the baseline model performs on `Dropout`, `Enrolled`, and `Graduate` separately.

```python
pd.DataFrame(metrics["Classification Report"]).transpose()
```
