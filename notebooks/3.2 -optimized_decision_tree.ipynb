# Optimized Decision Tree - Student Dropout Prediction

This notebook focuses on **Optimizing** the Decision Tree classifier using hyperparameter tuning and pruning techniques. We aim to improve the model's ability to generalize by preventing overfitting.

```python
import sys
from pathlib import Path

# Ensure the project root is in the Python path for utils import
PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import utils

# Ensure necessary directories exist
utils.ensure_directories()
```

## 1. Load Data

We load the same pre-processed dataset to ensure a fair comparison with the baseline.

```python
X_train, X_test, y_train, y_test = utils.load_processed_data()

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

## 2. Hyperparameter Optimization & Pruning

We will use `GridSearchCV` to explore optimal parameters. Additionally, we use **Cost Complexity Pruning (`ccp_alpha`)** which helps find the sub-tree that minimizes the cost-complexity function.

```python
param_grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [5, 10, 15, 20],
    "min_samples_split": [10, 20, 50],
    "min_samples_leaf": [5, 10, 20],
    "ccp_alpha": [0.0, 0.001, 0.01, 0.1],
    "random_state": [42]
}

dt_optimized = DecisionTreeClassifier(random_state=42)

metrics, summary_df = utils.train_and_evaluate(
    model=dt_optimized, 
    model_name="Decision Tree Optimized", 
    param_grid=param_grid
)
```

## 3. Results Overview

Let's view the best hyperparameters found and the performance of the optimized model.

```python
print("Best Parameters Found:")
for param, value in metrics["Best Parameters"].items():
    print(f" - {param}: {value}")

display(summary_df)
```
